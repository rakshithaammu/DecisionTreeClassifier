```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier

from sklearn.tree import export_graphviz,plot_tree
from IPython.display import Image
```

```python
df = pd.read_csv(r"C:\Users\raksh\Downloads\demodt.txt")
print(df.head())
```

```python
df.isnull().sum()
```

```python
feature = df[['Literacy','Cleanliness','Crime_Rate']]# Features - independent
Target = df['Good']# Target variable - dependent
```

```python
#Create Decision Tree classifier object
DT = DecisionTreeClassifier()
#Train Decision Tree Classifier
DT.fit(feature,Target)
```

```python
#Predict the response for test dataset
l=int(input("Enter Literacy: "))
cl=int(input("Enter Cleanliness: "))
cr=int(input("Crime_Rate: "))
pred = DT.predict([[l,cl,cr]])

if pred==1:
    print("Good State")
else: 
    print("Bad State")
```

```python
plot_tree(DT) #plot_tree(trained model name)
```
